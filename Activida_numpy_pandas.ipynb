## Ejercicio numpy

Imagina que tienes datos de temperatura promedio diaria de una ciudad durante un mes (30 días). Quieres realizar un análisis básico de estos datos usando Numpy para obtener información útil como la temperatura promedio del mes, la temperatura máxima y mínima, y los días en los que la temperatura estuvo por encima de la temperatura promedio.

1. **Crear un array de temperaturas:**

  - Genera un array de Numpy con 30 elementos que representen las temperaturas promedio diarias de la ciudad durante un mes.
  - Que se genere con numeros enteros aleatorios, entre los valores 18 y 29

2. **Calcular estadísticas básicas:**

  - Calcula la temperatura promedio del mes.
  - Encuentra la temperatura máxima y mínima del mes.

3. **Identificar días calurosos:**

  - Identifica los días en los que la temperatura estuvo por encima de la temperatura promedio del mes.

4. **Imprimir resultados:**

  - Imprime la temperatura promedio, máxima y mínima.
  - Imprime los índices (días) en los que la temperatura estuvo por encima del promedio.

```python
# Paso 1: Crear un array de temperaturas para 30 días
#temperaturas = np.array([25, 26, 27, 24, 23, 26, 28, 27, 25, 26, 24, 23, 25, 27, 28, 26, 27, 25, 24, 23, 26, 28, 27, 25, 26, 24, 23, 25, 27, 28])
temperaturas = np.array([25, 26, 27, 24, 23, 26, 28, 27, 25, 26, 24, 23, 25, 27, 28, 26, 27, 25, 24, 23, 26, 28, 27, 25, 26, 24, 23, 25, 27, 28])

temperaturas = np.random.randint(18,30,size=30, dtype=int)
print (temperaturas)

# Paso 2: Calcular estadísticas básicas
Temperatura_media = np.mean(temperaturas)
Temperatura_min = np.min(temperaturas)
Temperatura_max = np.max(temperaturas)


# Paso 3: Identificar días calurosos (con for o utilizar con función numpy wher
for días_calurosos in temperaturas:
    if días_calurosos >= 26:
        print(días_calurosos)
# np.where para encontrar los índices de los días en que la temperatura estuvo por encima del promedio
dias_calurosos = np.where  (temperaturas > Temperatura_media)[0]
print (dias_calurosos)
indices_calor = np.where(temperaturas >= 26)
print(indices_calor[0]+1)


# Paso 4: Imprimir resultados
print(f"Temperatura media: {Temperatura_media:.2f}")
print(f"Temperatura mínima: {Temperatura_min}")
print(f"Temperatura máxima: {Temperatura_max}")
print(f"Días calurosos por encima del promedio:{dias_calurosos + 1}")
```

### Ejercicio group by

1. Agrupar por `Product` y `Region`, y obtener la suma de las ventas
2. Filtrar el df para incluir solo aquellos productos donde las ventas son mayores que 1000, agrupar por `Product`, y obtener el promedio de las ventas para cada producto
3. Agrupar por `Region`, y obtener la suma de las ventas y el número total de ventas
4. Filtrar el df para incluir solo aquellas regiones donde la suma de ventas es mayor que 2000, agrupar por `Region`, y obtener la suma y el promedio de las ventas para cada región (usar lambda!!!)

```python
# Crear el df de ejemplo
data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': [1000, 1500, 800, 1200, 900, 1100, 750, 1300, 850, 950]
}

df = pd.DataFrame(data)
df.head(5)
```

```python
from ast import Expression
# 1: Agrupar por `Product` y `Region`, y obtener la suma de las ventas.
grouped_product_region = df.groupby(['Product', 'Region']).sum().reset_index()
print(grouped_product_region)
#otra forma
df.groupby(['Product', 'Region']).agg({"Sales" : 'sum'}).reset_index()

# 2: Filtrar el df para incluir solo aquellos productos donde las ventas son mayores que 1000,
# agrupar por `Product`, y obtener el promedio de las ventas para cada producto.
df[ (df['Sales']>1000) ].groupby(["Product"]).agg(
    {
        "Sales": ["mean"]
    }
).reset_index().round(2)
# 3: Agrupar por `Region`, y obtener la suma de las ventas y el número total de ventas.
grouped_region = df.groupby('Region').agg(
    {
        "Sales": ["sum", "count"]
    }
).reset_index()
print(grouped_region)

#otra forma
total_sales_region = df.groupby('Region').agg(Total_Sales=('Sales', 'sum'), Count_Sales=('Sales', 'count')).reset_index()
print (total_sales_region)

# 4: Filtrar el df para incluir solo aquellas regiones donde la suma de ventas es mayor que 2000,
# agrupar por `Region`, y obtener la suma y el promedio de las ventas para cada región.
df_fitrado = df.groupby("Region").filter(lambda x: x['Sales'].sum() > 2000 )
df_agrupado_4 = df.groupby("Region").filter(lambda x: x['Sales'].sum() > 2000 ).groupby("Region").agg(suma_ventas = ("Sales", "sum"),

                                 media_ventas = ("Sales", "mean")

                                 ).reset_index().round(2)

df_agrupado_4
```

```python
#2.Genero preferido por los hombres
hombres = df[df["Gender"] == "Male"]
genero_preferido_hombres = hombres["Genre"].value_counts().idxmax()
print(f"El genero preferido por los hombres es: {genero_preferido_hombres}")
```

```python
#3. Corroborar si los smartphones son el dispositivo en el cual permanecen más tiempo los usuarios mirando streaming
tipo_dispositivo = df[df['Device_Type'] == 'Smartphone']['Device_Type'].value_counts().idxmax()
print(f"El dispositivo en el cual permanecen más tiempo los usuarios mirando streaming es: {tipo_dispositivo}")
```

```python
#4. Las edades que suelen pagar más Premium
tipo_edad = df[df['Subscription_Status'] == 'Premium']['Age'].value_counts().idxmax()
print(f"Las edades que suelen pagar más Premium son: {tipo_edad}")
```

```python
premium_users = df[df["Subscription_Status"] == "Premium"]
edades_premium = premium_users["Age"].value_counts().sort_values(ascending=False)
edades_premium.head(1)
```

```python
#5.El video que tiene más tiempo de reproducción
video_mas_reproducido = df.groupby('Video_ID')['Duration_Watched (minutes)'].sum().idxmax()
print(f"El video que tiene más tiempo de reproducción es: {video_mas_reproducido}")
```

```python
#6. El género que más le gustó a las mujeres (rating)
genero_preferido_mujeres = df[df['Gender'] == 'Female']['Genre'].value_counts().idxmax()
print(f"El género que más le gustó a las mujeres es: {genero_preferido_mujeres}")
```

```python
#Conclusiones
print(f"Conclusiones:")

print(f"El genero preferido por los hombres es: {genero_preferido_hombres}")
print(f"El dispositivo en el cual permanecen más tiempo los usuarios mirando streaming es: {tipo_dispositivo}")
print(f"Las edades que suelen pagar más Premium son: {tipo_edad}")
print(f"El video que tiene más tiempo de reproducción es: {video_mas_reproducido}")
print(f"El género que más le gustó a las mujeres es: {genero_preferido_mujeres}")

```
